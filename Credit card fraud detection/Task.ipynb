```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('creditcard.csv')
df.head(5)
```

```python
# в данных нет проблем: все столбцы числовые, пропусков нет

df.info()
```

```python
df.describe()
```

```python
# классы несбалансированы

df.Class.value_counts() / len(df)
```

```python
# удалим повторяющиеся записи

df = df.drop_duplicates()
```

```python
# не очень понятно, что в данной задаче может привнести столбец со временем в подобном формате, поэтому избавимся от него

try:
    df = df.drop(['Time'], axis = 1)
except KeyError:
    pass
```

```python
# в данных явно нелинейные зависимости, так что можно попробовать применить линейную регрессию 
#или алгоритм, основанный на деревьях

sns.pairplot(df[['V1', 'V11', 'Amount', 'Class']], hue='Class')
```

```python
X = df.drop(['Class'], axis=1)
y = df.Class

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
# в качестве метрик будем рассматривать precision и recall
```

```python
pipe = Pipeline([('scaler', StandardScaler()), 
                 ('clf', LogisticRegression(class_weight = 'balanced', n_jobs=-1))])

#обучение модели
pipe.fit(Xtrain, ytrain)

#предсказание на тестовом наборе данных
ypred = pipe.predict(Xtest)

print(classification_report(ytest, ypred))
```

```python
#в итоге для первого класса плохие результаты
```

```python
pipe = Pipeline([('scaler', StandardScaler()), 
                 ('clf', RandomForestClassifier(class_weight = 'balanced', n_jobs=-1))])

pipe.fit(Xtrain, ytrain)

ypred = pipe.predict(Xtest)

print(classification_report(ytest, ypred))
```

```python
#теперь оба класса выровнялись по метрикам - будем использовать эту модель
#только попробуем сбалансировать классы
```

```python
from imblearn.pipeline import Pipeline 

pipe = Pipeline([('scaler', StandardScaler()), 
                 ('smote', SMOTE(random_state=42)),
                 ('clf', RandomForestClassifier(n_jobs=-1))])

pipe.fit(Xtrain, ytrain)

ypred = pipe.predict(Xtest)

print(classification_report(ytest, ypred))
```

```python
# precision после этого упал

# в качестве итоговой модели выберем рандомный лес без балансировки классов
```
